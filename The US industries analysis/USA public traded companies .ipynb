```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_1 = pd.read_csv("USA public companies.csv")
```

```python
file_2 = file_1.drop(["SEC Filings"], axis = 1)
file_2
```

```python
frequency_file = file_2["Sector"].value_counts()
```

```python
frequency_file.to_csv("frequency file.csv")
```

```python
new_f_file = pd.read_csv("frequency file.csv")
new_f_file.rename(columns={ new_f_file.columns[0]: "Sectors", new_f_file.columns[1]: "Frequency"}, inplace = True)
```

```python
new_f_file.to_csv("Sector freqeuncy file.csv")
```

```python
sector_frequency = pd.read_csv("Sector freqeuncy file.csv")
```

```python
sector_frequency.reset_index(drop = True, inplace = True)
sector_frequency
```

```python
sector_frequency.drop("Unnamed: 0", axis = 1)
```

```python
plt.bar(sector_frequency["Sectors"].unique(), sector_frequency["Frequency"])

plt.xticks(sector_frequency["Sectors"].unique(), rotation = "vertical", size = "10")
plt.ylabel("Frequency")
plt.xlabel("Sectors")
plt.show()
```
